```python
policy_ids = [101, 102, 103, 104, 105, 106]

claim_amounts = [5000, 15000, 7000, 20000, 3000, 12000]

claim_status = ["Approved", "Rejected", "Approved", "Approved", "Rejected", "Approved"]

customer_age = [25, 45, 35, 50, 23, 40]
```

## **Tasks**

Task 1: **Total Approved Claim Amount**


Calculate the total claim amount for all Approved claims.

```python
# Step - 1: Create a variable to store total_Amount
total_approved = 0

# Step 2: Loop row by row
for i in range(len(policy_ids)):
    if claim_status[i] == "Approved":   # filter condition
        total_approved += claim_amounts[i]

print("Total Approved Claim Amount:", total_approved)
```

Task 2: **Count Rejected Claims**

Find how many claims are Rejected.

```python
# step 1 : create a variable to count
Rejected_Claims = 0

# step 2: loop row by row
for i in range(len(policy_ids)) :
  if claim_status[i] == "Rejected":
    Rejected_Claims +=1
print(Rejected_Claims)
```
